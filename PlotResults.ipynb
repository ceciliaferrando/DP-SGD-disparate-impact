```python
import pandas as pd
import numpy as np
from scipy.special import softmax, expit
from scipy.stats import norm
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import log_loss
from sklearn.random_projection import GaussianRandomProjection
import matplotlib.pyplot as plt
```

```python
suffix = "_batch500_C4_noisescale0.5.npy"
suffix_base = "_batch500_C1_noisescale0.5.npy"
```

# Load results

```python
loss_vec = np.load("loss_vec" + suffix_base)
loss_vec_DP = np.load("loss_vec_DP" + suffix)

acc_train_vec = np.load("acc_train_vec" + suffix_base)
acc_train_vec_DP = np.load("acc_train_vec_DP" + suffix)

acc_test_vec = np.load("acc_test_vec" + suffix_base)
acc_test_vec_DP = np.load("acc_test_vec_DP" + suffix)

acc_o_train_vec = np.load("acc_o_train_vec" + suffix_base)
acc_o_train_vec_DP = np.load("acc_o_train_vec_DP" + suffix)

acc_o_test_vec = np.load("acc_o_test_vec" + suffix_base)
acc_o_test_vec_DP = np.load("acc_o_test_vec_DP" + suffix)

acc_8_train_vec = np.load("acc_8_train_vec" + suffix_base)
acc_8_train_vec_DP = np.load("acc_8_train_vec_DP" + suffix)

acc_8_test_vec = np.load("acc_8_test_vec" + suffix_base)
acc_8_test_vec_DP = np.load("acc_8_test_vec_DP" + suffix)
```

# Training losses

```python
plt.rcParams.update({'font.size': 22})
plt.figure(figsize=(12,8))
plt.plot(loss_vec, color = "blue", linewidth=4, label = "non-DP")
plt.plot(loss_vec_DP, color = "red", linewidth=4, label = "DP")
plt.xlabel("epochs")
plt.ylabel("train loss")
plt.legend()
plt.show()
```

# Train accuracy, non-DP/DP + balances/unbalanced

```python
plt.rcParams.update({'font.size': 22})
plt.figure(figsize=(12,8))
plt.plot(acc_o_train_vec, color = "forestgreen", linewidth=4, label = "non-DP, balanced")
plt.plot(acc_o_train_vec_DP, color = "blue", linewidth=4, label = "DP, balanced")
plt.plot(acc_8_train_vec, color = "lightgreen", linewidth=4, label = "non-DP, unbalanced")
plt.plot(acc_8_train_vec_DP, color = "cyan", linewidth=4, label = "DP, unbalanced")
plt.xlabel("epochs")
plt.ylabel("train accuracy")
plt.legend()
plt.show()
```

# Test accuracy, non-DP/DP + balances/unbalanced

```python
plt.rcParams.update({'font.size': 22})
plt.figure(figsize=(12,8))
plt.plot(acc_o_test_vec, color = "forestgreen", linewidth=4, label = "non-DP, balanced")
plt.plot(acc_o_test_vec_DP, color = "blue", linewidth=4, label = "DP, balanced")
plt.plot(acc_8_test_vec, color = "lightgreen", linewidth=4, label = "non-DP, unbalanced")
plt.plot(acc_8_test_vec_DP, color = "cyan", linewidth=4, label = "DP, unbalanced")
plt.xlabel("epochs")
plt.ylabel("test accuracy")
plt.legend()
plt.show()
```
